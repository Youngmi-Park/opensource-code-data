```python
import os
import re
import json
import tqdm
import pprint
import glob, os.path
import sys
from subprocess import Popen, PIPE
sys.setrecursionlimit(10**5)
```

```python
# 딕셔너리에서 키 중복 허용을 위한 클래스 정의
class file_data(object):
    def __init__(self,name):
        self.name = name

    def __str__(self):
        return self.name

    def __repr__(self):
        return "'"+self.name+"'"
    
# os.path.exists 메서드로 해당 Directory가 없으면 Directory를 생성
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

# 확장자를 확인해서 코드 파일 프로그래밍 언어 찾기
def find_language(extension):
    if extension == "md":
        return "markdown"
    elif extension == "py":
        return "python"
    elif extension == "java":
        return "java"
    elif extension == "c":
        return "c"
    elif extension in ["cc", "cpp"]:
        return "c++"
    elif extension == "cs":
        return "c#"
    elif extension == "js":
        return "javascript"
    elif extension == "html":
        return "html"  
    elif extension == "php":
        return "php"        
    else:
        return extension   

# 프로젝트 url 찾기
def find_repourl(base_path,user_name,project_name):
    oldpwd=os.getcwd()
    os.chdir(f"{base_path}/{user_name}/{project_name}")
    try:
        pipe = Popen("git remote -v", shell=True, stdout=PIPE)
        text = str(pipe.stdout.readlines()[0])
        repo_url = re.findall('(?:(?:https?|ftp):\/\/)?[\w/\-?=%.]+\.[\w/\-?=%.]+', text)
    finally:
        os.chdir(oldpwd)
    return str(repo_url[0])
```

```python
base_path = './opensourcedata/paperwithcode'
save_path = './opensourcedata_parsed/paperwithcode'
plist = os.listdir(base_path) # 폴더 안에 있는 user 리스트 가져옴
for i in tqdm.tqdm(plist):
    project_plist = os.listdir(os.path.join(base_path, i)) # user's project list
    for j in project_plist:
        # 1. 단순하게 프로젝트 폴더에서 파일만 읽어와서 순서대로 저장하는 형태
        flist = [x for x in glob.glob(f'{base_path}/{i}/{j}/**', recursive = True) if os.path.isfile(x)] # 프로젝트에서 파일만 받아오기
        dic ={}
        dic["user_name"] = i # user name
        dic["project_name"] = j # project name
        dic["contents"] = []
        #dic["repo_url"] = find_repourl(base_path,p_list[0]) # 프로젝트 url 주소
        for k in flist:
            ex = os.path.splitext(k)[1][1:] # extension
            filename = os.path.basename(k)
            if ex in ["py", "java", "c", "cc", "cpp", "cs", "js", "html", "php"]:
                file_dic = {}
                file_dic['type'] = 'code' # file type
                file_dic['file_name'] = filename # file name
                file_dic['language'] =  find_language(ex) # 확장자 확인
                file_dic['path'] = k # 파일 경로
                with open(k, 'r', encoding='utf-8') as f:
                    try:
                        fcontent = f.readlines() # zip 같은 경우 읽어지지 않음
                    except:
                        fcontent = None
                    file_dic['text'] = fcontent# 파일 내용
                dic["contents"].append(file_dic)
            elif ex in ["md"]:
                file_dic = {}
                file_dic['type'] = 'description'
                file_dic['file_name'] = filename # file name
                file_dic['language'] =  find_language(ex) # 확장자 확인
                file_dic['path'] = k # 파일 경로
                with open(k, 'r', encoding='utf-8') as f:
                    try:
                        fcontent = f.readlines() # zip 같은 경우 읽어지지 않음
                    except:
                        fcontent = None
                    file_dic['text'] = fcontent# 파일 내용
                dic["contents"].append(file_dic)
        # json 파일로 저장            
        createFolder(f"{save_path}/{i}")
        filesavepath = f"{save_path}/{i}/{j}.json"
        
       # newdic = clean_dict(dic)
        with open(filesavepath, 'w') as f:
            json.dump(dic, f, indent=4)
```
